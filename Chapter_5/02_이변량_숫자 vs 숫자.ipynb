## 3.ÏàòÏπòÌôî : ÏÉÅÍ¥ÄÎ∂ÑÏÑù

### (1) ÏÉÅÍ¥ÄÍ≥ÑÏàò, p-value

* ÏÉÅÍ¥ÄÍ≥ÑÏàò ùëü
    * Í≥µÎ∂ÑÏÇ∞ÏùÑ ÌëúÏ§ÄÌôî Ìïú Í∞í
    * -1 ~ 1 ÏÇ¨Ïù¥Ïùò Í∞í
    * -1, 1Ïóê Í∞ÄÍπåÏö∏ ÏàòÎ°ù Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÎÇòÌÉÄÎÉÑ.

* Í≤ΩÌóòÏóê ÏùòÌïú ÎåÄÎûµÏùò Í∏∞Ï§Ä(Ï†àÎåÄÏ†ÅÏù∏ Í∏∞Ï§ÄÏù¥ Ï†àÎåÄ ÏïÑÎãôÎãàÎã§.)
    * Í∞ïÌïú : 0.5 < |ùëü| ‚â§ 1
    * Ï§ëÍ∞Ñ : 0.2 < |ùëü| ‚â§ 0.5
    * ÏïΩÌïú : 0.1 < |ùëü| ‚â§ 0.2
    * (Í±∞Ïùò)ÏóÜÏùå : |ùëü| ‚â§ 0.1
    

```python
import scipy.stats as spst
```

```python
# ÏÉÅÍ¥ÄÍ≥ÑÏàòÏôÄ p-value 
spst.pearsonr(air['Temp'], air['Ozone'])
```

Í≤∞Í≥ºÎäî ÌäúÌîåÎ°ú ÎÇòÏò§ÎäîÎç∞

    * ÎåÄÎ¶ΩÍ∞ÄÏÑ§ : ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÍ∞Ä ÏûàÎã§.(ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä 0Ïù¥ ÏïÑÎãàÎã§.)

### (3) (Ï∂îÍ∞Ä) ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º heatmapÏúºÎ°ú ÏãúÍ∞ÅÌôî

```python
plt.figure(figsize = (8, 8))
sns.heatmap(air.corr(), 
            annot = True,            # Ïà´Ïûê(ÏÉÅÍ¥ÄÍ≥ÑÏàò) ÌëúÍ∏∞ Ïó¨Î∂Ä
            fmt = '.3f',             # Ïà´Ïûê Ìè¨Î©ß : ÏÜåÏàòÏ†ê 3ÏûêÎ¶¨ÍπåÏßÄ ÌëúÍ∏∞
            cmap = 'RdYlBu_r',       # ÏπºÎùºÎßµ
            vmin = -1, vmax = 1)     # Í∞íÏùò ÏµúÏÜå, ÏµúÎåÄÍ∞íÍ∞í
plt.show()
```
